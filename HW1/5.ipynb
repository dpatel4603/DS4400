## **Problem 5.1**

For each feature, write code to compute the average value, the min and max values, as well as its variance.
    
    Which features have the lowest and the highest average? Include the feature name and their average values for the features with the
    lowest and highest average.
    
    Which features have the lowest and the highest variance? Include the feature name and their variance values for the features with the
    lowest and highest variance.

```python
import pandas as pd

# Load the entire dataset
df = pd.read_csv('kc_house_data.csv')

# Drop columns to ignore: id, date (non-features), and zipcode (categorical)
df = df.drop(['id', 'date', 'zipcode'], axis=1)

# Separate response variable and features
y = df['price']  # Response (dependent variable)
X = df.drop('price', axis=1)  # Features (independent variables)
```

```python
# Compute statistics for each feature (excluding price - the response variable)
stats = pd.DataFrame({
    'Average': X.mean(),
    'Min': X.min(),
    'Max': X.max(),
    'Variance': X.var()
})

# Display the statistics table
stats
```

```python
# Find features with lowest and highest average
lowest_avg_feature = stats['Average'].idxmin()
highest_avg_feature = stats['Average'].idxmax()
lowest_avg_value = stats.loc[lowest_avg_feature, 'Average']
highest_avg_value = stats.loc[highest_avg_feature, 'Average']

print(f"the lowest average feature is {lowest_avg_feature} and the highest average feature is {highest_avg_feature}")
print(f"the lowest average value is {lowest_avg_value} and the highest average value is {highest_avg_value}")
```

## Features with Lowest and Highest Average

**Feature with LOWEST average:** `long` (longitude)  
**Average value:** -122.2139

**Feature with HIGHEST average:** `sqft_lot`  
**Average value:** 15,106.97

```python
# Find features with lowest and highest variance
lowest_var_feature = stats['Variance'].idxmin()
highest_var_feature = stats['Variance'].idxmax()
lowest_var_value = stats.loc[lowest_var_feature, 'Variance']
highest_var_value = stats.loc[highest_var_feature, 'Variance']

print(f"The lowest feature variance is {lowest_var_feature} and the highest feature variance is {highest_var_feature}")
print(f"The lowest variance value is {lowest_var_value} and the highest variance value is {highest_var_value}")
```

## Features with Lowest and Highest Variance

**Feature with LOWEST variance:** `waterfront`  
**Variance value:** 0.0075

**Feature with HIGHEST variance:** `sqft_lot`  
**Variance value:** 1,715,658,774.18 (approximately 1.72 billion)

## **Problem 5.2:** 

Compute the correlation coefficient of each feature with the response. Include a table with the correlation coefficient of each feature with the response. 

Which features are positively correlated (i.e., have positive correlation coefficient) with the response?

Which feature has the highest positive correlation with the response?

```python
# Compute correlation coefficient of each feature with the response (price)
correlations = X.corrwith(y)

# Create a table sorted by correlation value
corr_table = pd.DataFrame({
    'Correlation with Price': correlations
}).sort_values(by='Correlation with Price', ascending=False)

corr_table
```

```python
# Find features positively correlated with price
positive_corr = correlations[correlations > 0].sort_values(ascending=False)

# Feature with highest positive correlation
highest_corr_feature = positive_corr.idxmax()
highest_corr_value = positive_corr.max()
```
