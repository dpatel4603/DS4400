Problem 4.2: Generate at random 3 matrices of size $3 \times 3$ and fill each entry with a random integer chosen from -10 to 10. Use an existing package to compute the inverse of each matrix if it exists. Include in your report:

    - Each of the 3 matrices
    
    - The inverses computed with the package
    
    - Compute the product of each matrix with its inverse to check that you obtain the identity matrix.

```python
import numpy as np
```

```python
rng = np.random.default_rng(777)
```

```python
def random_matrix():
    return rng.integers(low=-10, high=11, size=(3, 3))

def show_matrix_blcok(i, A):
    print(f"\n MATRIX {i}")
    print(A)
    
    det = round(np.linalg.det(A))
    print(f"det(A) approx {det}")
    
    A_inv = np.linalg.inv(A)
    print(f"A_inv approx {A_inv}")
    
    prod = A @ A_inv
    print(f"]nA @ A_inv should be identity: {prod}")
    
    print("\nCheck allclose to identity:", np.allclose(prod, np.eye(3)))
        

        
```

```python
for i in range(1, 4):
    A = random_matrix()
    show_matrix_blcok(i,A)
```
